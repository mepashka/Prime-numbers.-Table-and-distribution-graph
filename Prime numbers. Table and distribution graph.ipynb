```python
import math
import matplotlib.pyplot as plt
import numpy as np

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True

def prime_numbers(n):
    primes = []
    for num in range(1, n + 1):
        if is_prime(num):
            primes.append(num)
    return primes

def plot_color_curve(primes):
    n = len(primes)

    fig, ax = plt.subplots()
    colors = range(n)

    for i in range(n):
        ax.plot(i, primes[i], marker='o', markersize=8, color=plt.cm.jet(colors[i]/n), markeredgecolor='black')

    ax.set_xlabel('Индекс')
    ax.set_ylabel('Простое число')
    ax.set_title('Кривая с полигональной раскраской для распределения простых чисел')

    plt.show()

def main():
    n = int(input("Введите число: "))
    primes = prime_numbers(n)
    plot_color_curve(primes)

if __name__ == "__main__":
    main()
```
