```python
import math
import matplotlib.pyplot as plt
```

```python
def is_prime(num):
    # Проверяем, является ли число простым
    if num < 2:
        return False
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True
```

```python
def prime_numbers(n):
    # Находим все простые числа от 1 до n
    primes = []
    for num in range(1, n + 1):
        if is_prime(num):
            primes.append(num)
    return primes
```

```python
def plot_color_curve(primes):
    # Строим кривую с полигональной раскраской для простых чисел
    n = len(primes)

    fig, ax = plt.subplots()
    colors = range(n)

    for i in range(n):
        ax.plot(colors[i], primes[i], marker='o', markersize=8, color=plt.cm.jet(i/n), markeredgecolor='black')

    ax.set_xlabel('Индекс')
    ax.set_ylabel('Простое число')
    ax.set_title('Кривая с полигональной раскраской для распределения простых чисел')
    ax.set_yticks(primes)
    ax.set_ylim(min(primes)-1, max(primes)+1)

    plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)  # Изменение отступов

    plt.show()
```

```python
def print_prime_table(primes):
    # Выводим список простых чисел в виде таблицы
    n = len(primes)
    columns = 5

    print("Список простых чисел:")
    for i in range(0, n, columns):
        row = primes[i:i+columns]
        row_str = " ".join(str(num) for num in row)
        print(row_str)
```

```python
def main():
    # Основная функция программы
    n = int(input("Введите число: "))
    primes = prime_numbers(n)
    plot_color_curve(primes)
    print_prime_table(primes)
```

```python
if __name__ == "__main__":
    main()
```
